## Read File

```python
def read_txt_file(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()
    return [line.strip() for line in lines]

file_path = 'TESTCARD.txt'  # replace with your file path
lines = read_txt_file(file_path)
print(lines)  # Display the content of the file
```

## Create Widgets

```python
def create_widgets(lines):
    widgets_dict = {}
    for idx, line in enumerate(lines):
        try:
            number = float(line)  # Try converting line to a float
            widgets_dict[idx] = widgets.FloatText(value=number, description=f'Line {idx + 1}')
        except ValueError:
            widgets_dict[idx] = widgets.Text(value=line, description=f'Line {idx + 1}')  # Keep it as text if not a number
    return widgets_dict

widgets_dict = create_widgets(lines)
```

## Display 

```python
for widget in widgets_dict.values():
    display(widget)
```

## Write over .txt file

```python
def update_txt_file(file_path, widgets_dict):
    with open(file_path, 'w') as file:
        for widget in widgets_dict.values():
            file.write(str(widget.value) + '\n')

update_button = widgets.Button(description="Update File")

def on_update_button_clicked(b):
    update_txt_file(file_path, widgets_dict)
    print("File updated!")

update_button.on_click(on_update_button_clicked)
display(update_button)
```
